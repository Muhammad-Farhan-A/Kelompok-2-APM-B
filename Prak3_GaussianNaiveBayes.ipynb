```python
#import library
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset
dataset = pd.read_csv("heart_dataset.csv")
dataset.head()
```

```python
#analisis dataset
dataset.info()
```

```python
#variabel independen
x = dataset.drop(['target'], axis = 1)
x.head()
```

```python
#variabel dependen
y = dataset['target']
y.head()
```

```python
#split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

```python
#mengaktifkan package dan fungsi gaussian naive bayes
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
#memasukkan data training pada fungsi
nb = model.fit(x_train, y_train)
```

```python
#melakukan prediksi
y_pred = nb.predict(x_test)
y_pred
```

```python
#menentukan probabilitas hasil prediksi
nb.predict_proba(x_test)
```

```python
#akurasi
print("Gaussian Naive Bayes score: ",nb.score(x_test, y_test))
```
